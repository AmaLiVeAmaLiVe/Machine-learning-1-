# Řešení

## I. Příprava dat

Importuju potřebné balíčky

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

Načítam si dataset ze csv-souboru a podívám se na informace o něm

````python
df = pd.read_csv('data.csv')
display(df.shape)
display(df.head())
display(df.info())
display(df.nunique())
display(df.describe())
````

````python
df.columns
````

Vytvořím nový sloupec délky jmen

````python
len_names = []

for name in df['name']:
    len_names.append(len(name))

df['len_names'] = len_names
````

Nahrádím chybějící data v sloupci věku průměrem veku

````python
df['age'] = df['age'].fillna(df['age'].mean())
````

Nahrádím chybějící data ve všem datasetu -1

````python
df = df.replace(np.NaN, -1)
````

Odstraním nepotřebné sloupce

````python
df = df.drop(['ID', 'name', 'ticket'], axis=1)
````

Převodím všechny slopuce typu 'object' na numerický typ

````python
string_cols = df.select_dtypes(['object']).columns
df[string_cols] = df[string_cols].astype('category').apply(lambda x: x.cat.codes)
display(df.head())
display(df.info())
````

Vidím, že nyní nemáme žádná chybějící data v našem datasetu. Tedy, náš dataset je už přípraven na vložení do modelů.

````python
df.isnull().sum(axis=0)
````

Udělám vizualizaci pro korelace různých příznaků s mírou přežití

````python
fig, axes = plt.subplots(2, 2, figsize=(20,10))
fig.suptitle('Surviving stats by distinct features')

sns.countplot(ax=axes[0,0], data=df, x="sex", hue="survived", edgecolor = "black", palette='husl')
axes[0, 0].set_title('Survivng stats by sex')
sns.countplot(ax=axes[0,1], data=df, x="embarked", hue="survived", edgecolor = "black", palette='husl')
axes[0, 1].set_title('Survivng stats by embarked')
sns.countplot(ax=axes[1,0], data=df, x="pclass", hue="survived", edgecolor = "black", palette='husl')
axes[1, 0].set_title('Survivng stats by pclass')
sns.countplot(ax=axes[1,1], data=df, x="parch", hue="survived", edgecolor = "black", palette='husl')
axes[1, 1].set_title('Survivng stats by parch')

plt.show()
````

## II. Použítí modelů machine learningu

### II(a). Rozdělení dat na trénovací, validační a testovací množiny

Nejperve, rozdělíme si dataset na cílovou proměnnou (v našem případě to je 'survived'), kterou chceme predikovat, a na matici příznaků, pomocí kterou budeme predikovat.

````python
from sklearn.model_selection import train_test_split

Xdata = df.drop('survived', axis=1)
ydata = df['survived']
````

Ted' si už můžeme rozdělit data na trénovací, validační a testovací množiny.

````python
rd_seed = 42

Xtrain, Xtest, ytrain, ytest = train_test_split(Xdata, ydata, test_size=0.4, random_state=rd_seed) 
display(Xtrain.shape)
display(ytrain.shape)
display(Xtest.shape)
display(ytest.shape)
````

````python
Xval, Xtest, yval, ytest = train_test_split(Xtest, ytest, test_size=0.5, random_state=rd_seed) 
display(Xval.shape)
display(yval.shape)
display(Xtest.shape)
display(ytest.shape)
````

Dostali jsme 60% trénovacích dat, 20% na validaci naších modelů a zbytek - 20% na jejich testování 

### II(b). Implementace a použití modelů DecisionTreeClassifier

Implementiruju modele DecisionTreeClassifier s parametry hloubky 1-50 a kritéria Entropie a Gini Index

````python
#postup byl inspirovan postupem z notebooku cviceni2

from sklearn.model_selection import ParameterGrid
from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics as metrics
from sklearn.metrics import roc_curve, roc_auc_score 

param_grid = {
    'max_depth': range(1,50), 
    'criterion': ['entropy', 'gini']
}
param_comb = ParameterGrid(param_grid)

val_acc = []
train_acc = []

for params in param_comb:
    clf_tree = DecisionTreeClassifier(max_depth=params['max_depth'], criterion=params['criterion'])
    clf_tree.fit(Xtrain, ytrain)
    train_acc.append(metrics.accuracy_score(ytrain, clf_tree.predict(Xtrain)))
    val_acc.append(metrics.accuracy_score(yval, clf_tree.predict(Xval)))
````

Ted provádím vizulizace pro poměr metrik přesnosti na validačních datech k metrikám přesnosti na trénovacíh datech

````python
#postup byl inspirovan postupem z notebooku cviceni2

plt.figure(figsize=(20,6))
plt.plot(train_acc,'or-')
plt.plot(val_acc,'ob-')
plt.xlabel('hyperparametr index')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])
````

Hledám nejlepší parametry pro DecisionTreeClassifier

````python
best_params_tree = param_comb[np.argmax(val_acc)]
print(np.max(val_acc))
print(f'Best parameters for DecisionTreeClassifier - {best_params_tree}')
````

Počítám f1 score, ROC AUC score a kreslím křívku ROC pro model s nejlepšími parametry

````python
clf_best_tree = DecisionTreeClassifier(**best_params_tree)
clf_best_tree.fit(Xtrain, ytrain)
y_pred = clf_best_tree.predict(Xval)
y_pred_proba = clf_best_tree.predict_proba(Xval)

print('F1 score (validation) for DecisionTreeClassifier: {0:.6f}'.format(metrics.f1_score(yval, y_pred)))
````

````python
roc_auc = metrics.roc_auc_score(yval, y_pred_proba[:,1])
print(f'ROC AUC score (validation) for DecisionTreeClassifier {roc_auc}')
````

````python
fpr, tpr, thresholds = roc_curve(yval, y_pred_proba[:,1], pos_label=1)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc) 
plt.plot([0, 1], [0, 1], 'k--', label='Random classifier') 
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.title('ROC Curve') 
plt.legend(loc="lower right") 
plt.show()
````

Vidím, že mám dost dobré metriky F1 score and AUC ROC při použití DecisionTreeClassifier pro náš dataset.

### II(c). Implementace a použití modelů kNN

Implementiruju modele DecisionTreeClassifier s parametry poučtu sousedů 1-35 a weights: uniform a distance. Nejperve normalizuju data pro další použití.

Normalizuju data pomocí standatizace.

````python
#postup byl inspirovan postupem z notebooku cviceni4

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xtrain_standard = scaler.fit_transform(Xtrain)
Xval_standard = scaler.transform(Xval)

param_grid = {
    'n_neighbors': range(1,35), 
    'weights': ['uniform', 'distance']
}
param_comb = ParameterGrid(param_grid)

val_acc = []
train_acc = []

for params in param_comb:
    clf_knn = KNeighborsClassifier(n_neighbors=params['n_neighbors'], weights=params['weights'])
    clf_knn.fit(Xtrain_standard, ytrain)
    val_acc.append(clf_knn.score(Xval_standard, yval))
    train_acc.append(clf_knn.score(Xtrain_standard, ytrain))

print(f"Validation accuracy of normalized model: {clf_knn.score(Xval_standard, yval):0.4f}")
````

Ted zkusíme min-max normalizaci dat.

````python
#postup byl inspirovan postupem z notebooku cviceni4

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
Xtrain_min_max = scaler.fit_transform(Xtrain)
Xval_min_max = scaler.transform(Xval)

val_acc = []
train_acc = []

for params in param_comb:
    clf_knn = KNeighborsClassifier(n_neighbors=params['n_neighbors'], weights=params['weights'])
    clf_knn.fit(Xtrain_min_max, ytrain)
    val_acc.append(clf_knn.score(Xval_min_max, yval))
    train_acc.append(clf_knn.score(Xtrain_min_max, ytrain))

print(f"Validation accuracy of normalized model: {clf_knn.score(Xval_min_max, yval):0.4f}")
````

Vidím, že přesnost při použití min-max normalizaci dat je trochu lepší. Náší volbou tedy je druhá možnost

Ted provádím vizulizace pro poměr metrik přesnosti na validačních datech k metrikám přesnosti na trénovacíh datech

````python
#postup byl inspirovan postupem z notebooku cviceni2

plt.figure(figsize=(20,6))
plt.plot(train_acc,'or-')
plt.plot(val_acc,'ob-')
plt.xlabel('hyperparametr index')
plt.ylabel('accuracy')
plt.legend(['train', 'validation'])
````

Hledám nejlepší parametry pro kNN

````python
best_params_knn = param_comb[np.argmax(val_acc)]
print(np.max(val_acc))
print(f'Best parameters for kNN - {best_params_knn}')
````

Počítám f1 score, ROC AUC score a kreslím křívku ROC pro model s nejlepšími parametry

````python
clf_best_knn = KNeighborsClassifier(**best_params_knn)
clf_best_knn.fit(Xtrain_min_max, ytrain)
y_pred = clf_best_knn.predict(Xval_min_max)
y_pred_proba = clf_best_knn.predict_proba(Xval_min_max)

print('F1 score (validation) for kNN method: {0:.6f}'.format(metrics.f1_score(yval, y_pred)))
````

````python
roc_auc = roc_auc_score(yval, y_pred_proba[:,1]) 
print(f'ROC AUC score (validation) for kNN method {roc_auc}')
````

````python
fpr, tpr, thresholds = roc_curve(yval, y_pred_proba[:,1], pos_label=1)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc) 
plt.plot([0, 1], [0, 1], 'k--', label='Random classifier') 
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.title('ROC Curve') 
plt.legend(loc="lower right") 
plt.show()
````

Vidím, že kNN metoda má skoro stejné metriky: AUC ROC je téměř stejný, ale F1 score je trochu horší.

## III. Finální model

Vzhledem k metrikám obou modelů mojí volbou pro finální model je tedy DecisionTreeClassifier s následující parametry

````python
print(f'Parameters of the final model - {best_params_tree}')
````

Zkusím ted testovací data na našem modelu. Počítám F1 score, ROC AUC score a kreslím ROC křívku pro finální model s testovacími daty

````python
clf_final = DecisionTreeClassifier(**best_params_tree)
clf_final.fit(Xtrain, ytrain)
y_pred = clf_final.predict(Xtest)
y_pred_proba = clf_final.predict_proba(Xtest)

print('Accuracy score (test) for final model: {0:.6f}'.format(metrics.accuracy_score(ytest, y_pred)))
print('F1 score (test) for final model: {0:.6f}'.format(metrics.f1_score(ytest, y_pred)))
````

````python
roc_auc = metrics.roc_auc_score(ytest, y_pred_proba[:,1])
print(f'ROC AUC score (test) for final model {roc_auc}')
````

````python
fpr, tpr, thresholds = roc_curve(ytest, y_pred_proba[:,1], pos_label=1)

plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc) 
plt.plot([0, 1], [0, 1], 'k--', label='Random classifier') 
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.title('ROC Curve') 
plt.legend(loc="lower right") 
plt.show()
````

Ze získaných metrik na testovacích datech můžu říct, že můj zvolený model dává dost dobré výsledky.
